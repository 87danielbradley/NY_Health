```python
import csv
import pandas as pd
import numpy as np
import os
import json
```

```python
file_path="Community_Health_Indicator_Reports__CHIRS___Latest_Data.csv"

health_stats=pd.read_csv(file_path)
```

```python
health_stats.head()
```

```python
obs_stats=health_stats[health_stats['Indicator']=='Age-adjusted percentage of adults obese (BMI 30 or higher)']
```

```python
obs_stats.head()
```

```python
obs_stats=obs_stats[['County Name','Indicator','Percent/Rate','Location']]
obs_stats=obs_stats.rename(columns={"County Name": "County", "Indicator": "Health_Topic","Percent/Rate":"Percentage"})
obs_stats.head()
```

```python
obs_stats
```

```python
obs_stats=obs_stats[obs_stats['Location'].notna()]
obs_stats.shape
```

```python
obs_stats
```

```python
NY_stats=obs_stats.to_json(orient='records')

with open('NY_Obs_Stats.json', 'w') as f:
    json.dump(NY_stats, f, indent=2)
```

```python
geo_coord=obs_stats[['County','Location']]
geo_json=geo_coord.to_json(orient='records')

with open('Geo_Loc_Counties.json', 'w') as f:
    json.dump(geo_json, f, indent=2)
```
