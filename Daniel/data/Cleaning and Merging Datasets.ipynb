```python
import pandas as pd
```

```python
data2010 = pd.read_csv('data/cleaning2010v2.csv')
data2011 = pd.read_csv('data/cleaning2011v2.csv')
data2012 = pd.read_csv('data/cleaning2012v2.csv')
data2013 = pd.read_csv('data/cleaning2013v2.csv')
data2014 = pd.read_csv('data/cleaning2014v2.csv')
data2015 = pd.read_csv('data/cleaning2015v2.csv')
data2016 = pd.read_csv('data/cleaning2016v2.csv')
data2017 = pd.read_csv('data/cleaning2017v2.csv')
data2018 = pd.read_csv('data/cleaning2018v2.csv')
```

```python

race=['total', 'white', 'black', 'native', 'asian', 'islander','other','mixed','latino']
fullLoD = []
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018]
datasets=[data2010,data2011,data2012,data2013,data2014,data2015,data2016,data2017,data2018]
for j in range(0,len(year)):
    data = datasets[j]
    pwhite = data['White Percentage']
    pwhitemoe =data['Margin of Error White Percentage']
    pblack =data['Black Percentage']
    pblackmoe = data['Margin of Error Black Percentage']
    pnative = data['Native Percentage']
    pnativemoe = data['Margin of Error Native Percentage']
    pasian = data['Asian Percentage']
    pasianmoe = data['Margin of Error Asian Percentage']
    pother = data['Other Percentage']
    pothermoe = data['Margin of Error Other Percentage']
    pislander = data['Islander Percentage']
    pislandermoe = data['Margin of Error Islander Percentage']
    pmixed = data['Mixed Percentage']
    pmixedmoe = data['Margin of Error Mixed Percentage']
    platino =  data['Latino Total Percentage']
    platinomoe =  data['Margin of Error Latino Total Percentage']
    ptotal = 100
    ptotalmoe = data['Margin of Error Total population']
                          
    counties = data['Geographic Area Name']
    meanTotal = data['Mean Income Total']
    meanTotalMOE = data['Margin of Error Mean Income Total']
    meanWhite = data['Mean Income White']
    meanWhiteMOE = data['Margin of Error Mean Income White']
    meanBlack = data['Mean Income Black']
    meanBlackMOE = data['Margin of Error Mean Income Black']
    meanNative = data['Mean Income Native']
    meanNativeMOE = data['Margin of Error Mean Income Native']
    meanAsian = data['Mean Income Asian']
    meanAsianMOE = data['Margin of Error Mean Income Asian']
    meanIslander = data['Mean Income Islander']
    meanIslanderMOE = data['Margin of Error Mean Income Islander']
    meanOther = data['Mean Income Other'] 
    meanOtherMOE = data['Margin of Error Mean Income Other']
    meanMixed = data['Mean Income Mixed']
    meanMixedMOE = data['Margin of Error Mean Income Mixed']
    meanLatino = data['Mean Income Latino Total']
    meanLatinoMOE = data['Margin of Error Mean Income Latino Total']
    meanWhiteAlone = data['Mean Income White Alone']
    meanWhiteAloneMOE = data['Margin of Error Mean Income White Alone']    
    meanIncome=[meanTotal,meanWhite,meanBlack,meanNative,meanAsian,meanIslander,meanOther,meanMixed,meanLatino]
    percentage = [ptotal,pwhite,pblack,pnative,pasian,pislander,pother,pmixed,platino]
    
    for i in range(0,len(race)):
        myFullDict = {'race': race[i],
                  'date':year[j],
                  'meanIncome':meanIncome[i],
                  'county':counties,
                  'percentage':percentage[i]}
        fullLoD.append(myFullDict)
```

```python
output_df = pd.DataFrame(fullLoD[0])
for i in range(1,len(fullLoD)):
    df2= pd.DataFrame(fullLoD[i])
    output_df = pd.concat([output_df,df2])
output_df=output_df.reset_index(drop=True)

output_df.head()
```

```python
flag_count = 0
for i in range(0,len(output_df["meanIncome"])):
    try:
        int(output_df['meanIncome'][i])
    except:
        flag_count+=1
        output_df['meanIncome'][i]=0
    try:
        int(output_df['percentage'][i])
    except:
        flag_count+=1
        output_df['percentage'][i]=0
output_df
```

```python
output_df.to_csv('popandincome.csv',index=False)
```

```python

race=['total', 'white', 'black', 'native', 'asian', 'islander','other','mixed','latino','whiteAlone']
LoD = []
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018]
datasets=[data2010,data2011,data2012,data2013,data2014,data2015,data2016,data2017,data2018]
for j in range(0,len(year)):
    data = datasets[j]
    data = data.drop(columns=['id',
                           'White Percentage','Margin of Error White Percentage',
                           'Black Percentage','Margin of Error Black Percentage',
                           'Native Percentage','Margin of Error Native Percentage',
                           'Asian Percentage','Margin of Error Asian Percentage',
                           'Other Percentage','Margin of Error Other Percentage',
                           'Islander Percentage','Margin of Error Islander Percentage',
                           'Mixed Percentage', 'Margin of Error Mixed Percentage',
                           'Latino Total Percentage', 'Margin of Error Latino Total Percentage',
                           'White Alone Percentage', 'Margin of Error White Alone Percentage',
                           'Total population', 'Margin of Error Total population'
                          ])
    counties = data['Geographic Area Name']
    meanTotal = data['Mean Income Total']
    meanTotalMOE = data['Margin of Error Mean Income Total']
    meanWhite = data['Mean Income White']
    meanWhiteMOE = data['Margin of Error Mean Income White']
    meanBlack = data['Mean Income Black']
    meanBlackMOE = data['Margin of Error Mean Income Black']
    meanNative = data['Mean Income Native']
    meanNativeMOE = data['Margin of Error Mean Income Native']
    meanAsian = data['Mean Income Asian']
    meanAsianMOE = data['Margin of Error Mean Income Asian']
    meanIslander = data['Mean Income Islander']
    meanIslanderMOE = data['Margin of Error Mean Income Islander']
    meanOther = data['Mean Income Other'] 
    meanOtherMOE = data['Margin of Error Mean Income Other']
    meanMixed = data['Mean Income Mixed']
    meanMixedMOE = data['Margin of Error Mean Income Mixed']
    meanLatino = data['Mean Income Latino Total']
    meanLatinoMOE = data['Margin of Error Mean Income Latino Total']
    meanWhiteAlone = data['Mean Income White Alone']
    meanWhiteAloneMOE = data['Margin of Error Mean Income White Alone']    
    meanIncome=[meanTotal,meanWhite,meanBlack,meanNative,meanAsian,meanIslander,meanOther,meanMixed,meanLatino,meanWhiteAlone]
    MOE = [meanTotalMOE,meanWhiteMOE,meanBlackMOE,meanNativeMOE,meanAsianMOE,meanIslanderMOE,meanOtherMOE,meanMixedMOE,meanLatinoMOE,meanWhiteAloneMOE]

    
    for i in range(0,len(race)):
        myDict = {'race': race[i],
                  'date':year[j],
                  'meanIncome':meanIncome[i],
                  'county':counties,
                  'MOE':MOE[i]}
        LoD.append(myDict)
```

```python
output_df = pd.DataFrame(LoD[0])
for i in range(1,len(LoD)):
    df2= pd.DataFrame(LoD[i])
    output_df = pd.concat([output_df,df2])
output_df=output_df.reset_index(drop=True)

output_df.head()
```

```python
flag_count = 0
for i in range(0,len(output_df["meanIncome"])):
    try:
        int(output_df['meanIncome'][i])
    except:
        flag_count+=1
        output_df['meanIncome'][i]=0
    try:
        int(output_df['MOE'][i])
    except:
        flag_count+=1
        output_df['MOE'][i]=0
output_df
```

```python
output_df.to_csv('compiledData.csv',index=False)
```
