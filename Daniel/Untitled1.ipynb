```python
with open('test.geojson','w') as f:
    for i in range(0,len(temp_df)):
        try:
            perCapitaTotal=int(temp_df.perCapitaTotal[i])
        except:
            perCapitaTotal="NaN"
            
        try:
            meanIncomeTotal=int(temp_df.meanIncomeTotal[i])
        except: 
            meanIncomeTotal='NaN'
        try:
            perCapitaWhite=int(temp_df.perCapitaWhite[i])
        except: 
            perCapitaWhite='NaN'
        try:
            meanIncomeWhite=int(temp_df.meanIncomeWhite[i])
        except: 
            meanIncomeWhite='NaN'
        try:
            perCapitaBlack=int(temp_df.perCapitaBlack[i])
        except: 
            perCapitaBlack='NaN'
        try:
            meanIncomeBlack=int(temp_df.meanIncomeBlack[i])
        except: 
            meanIncomeBlack='NaN'
        try:
            perCapitaNative=int(temp_df.perCapitaNative[i])
        except: 
            perCapitaNative='NaN'
        try:
            meanIncomeNative=int(temp_df.meanIncomeNative[i])
        except: 
            meanIncomeNative='NaN'
        try:
            perCapitaAsian=int(temp_df.perCapitaAsian[i])
        except: 
            perCapitaAsian='NaN'
        try:
            meanIncomeAsian=int(temp_df.meanIncomeAsian[i])
        except: 
            meanIncomeAsian='NaN'
        try:
            perCapitaIslander=int(temp_df.perCapitaIslander[i])
        except: 
            perCapitaIslander='NaN'
        try:
            meanIncomeIslander=int(temp_df.meanIncomeIslander[i])
        except: 
            meanIncomeIslander='NaN'
        try:
            perCapitaOther=int(temp_df.perCapitaOther[i])
        except: 
            perCapitaOther='NaN'
        try:
            meanIncomeOther=int(temp_df.meanIncomeOther[i])
        except: 
            meanIncomeOther='NaN'
        try:
            perCapitaTwoRaces=int(temp_df.perCapitaTwoRaces[i])
        except: 
            perCapitaTwoRaces='NaN'
        try:
            meanIncomeTwoRaces=int(temp_df.meanIncomeTwoRaces[i])
        except: 
            meanIncomeTwoRaces='NaN'
        try:
            perCapitaTotalLatino=int(temp_df.perCapitaTotalLatino[i])
        except: 
            perCapitaTotalLatino='NaN'
        try:
            meanIncomeTotalLatino=int(temp_df.meanIncomeTotalLatino[i])
        except: 
            meanIncomeTotalLatino='NaN'
        try:
            perCapitaWhitealone=int(temp_df.perCapitaWhitealone[i])
        except: 
            perCapitaWhitealone='NaN'
        try:
            meanIncomeWhitealone=int(temp_df.meanIncomeWhitealone[i])
        except: 
            meanIncomeWhitealone='NaN'

        
        point = Point((temp_df.latitude[i],temp_df.longitude[i]))
        features = []
        features.append(Feature(geometry=point, properties = {"county":temp_df.location[i].split(", ")[0],
                                                              "state":temp_df.location[i].split(", ")[1],
                                                              "perCapitaTotal":perCapitaTotal,
        
                                                              "meanIncomeTotal":meanIncomeTotal,
                                                              "perCapitaWhite":perCapitaWhite,
                                                              "meanIncomeWhite":meanIncomeWhite,
                                                              "perCapitaBlack":perCapitaBlack,
                                                              "meanIncomeBlack":meanIncomeBlack,
                                                              "perCapitaNative":perCapitaNative,
                                                              "meanIncomeNative":meanIncomeNative,
                                                              "perCapitaAsian":perCapitaAsian,
                                                              "meanIncomeAsian":meanIncomeAsian,
                                                              "perCapitaIslander":perCapitaIslander,
                                                              "meanIncomeIslander":meanIncomeIslander,
                                                              "perCapitaOther":perCapitaOther,
                                                              "meanIncomeOther":meanIncomeOther,
                                                              "perCapitaTwoRaces":perCapitaTwoRaces,
                                                              "meanIncomeTwoRaces":meanIncomeTwoRaces,
                                                              "perCapitaTotalLatino":perCapitaTotalLatino,
                                                              "meanIncomeTotalLatino":meanIncomeTotalLatino,
                                                              "perCapitaWhitealone":perCapitaWhitealone,
                                                              "meanIncomeWhitealone":meanIncomeWhitealone
                                                             }))
        print(i)
        feature_collection = FeatureCollection(features)

        dump(feature_collection, f, indent=2)
```
